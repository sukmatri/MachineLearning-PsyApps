```python
import pandas as pd
```

```python
dataset3 = pd.read_csv('dataset_mental2.csv')
```

```python
dataset3
```

```python
dataset3 = dataset3.drop('Unnamed: 0', axis = True)
```

```python
dataset3
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_dataset3 = scaler.fit_transform(dataset3)
print(scaled_dataset3)
```

```python
from sklearn.cluster import KMeans
scaled_dataset3 = pd.DataFrame(scaled_dataset3, columns =['Age', 'Gender', 'family_history', 'treatment', 'work_interfere', 'care_options', 'wellness_program', 'seek_help', 'mental_health_consequence', 'phys_health_consequence'])

scaler = StandardScaler()
scaled_features = scaler.fit_transform(scaled_dataset3)

kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(scaled_features)

scaled_dataset3['Cluster'] = kmeans.labels_

print(scaled_dataset3)
```

```python
dataset3['Cluster'] = scaled_dataset3['Cluster'].values
```

```python
dataset3
```

```python
fix = pd.DataFrame(dataset3)
```

```python
fix.to_csv('dataset_mental3.csv')
```
