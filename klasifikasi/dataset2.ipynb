```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
scaler = StandardScaler()
dataset2 = scaler.fit_transform(dataset2)
print(dataset2)
```

```python
from sklearn.cluster import KMeans
dataset2 = pd.DataFrame(dataset2, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature7', 'Feature8', 'Feature9', 'Feature10', 'Feature6'])

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(dataset2)

# Apply K-Means with 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(scaled_features)

# Add cluster labels to the original DataFrame
dataset2['Cluster'] = kmeans.labels_

# Display the DataFrame with cluster labels
print(dataset2)
```
